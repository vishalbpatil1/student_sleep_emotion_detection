```python
import tensorflow as tf 
import cv2
import time
import pandas as pd
import numpy as np
from keras.preprocessing import image

model1=tf.keras.models.load_model(r'C:\\Users\\Vishal\\face_expression_recognition\\modelface_network1.h5')


#cap=cv2.VideoCapture(r"C:\\Users\\Vishal\\New_idea\\demo_video.mp4")
#cap=cv2.VideoCapture(r"C:\\Users\\Vishal\\Student_sleep\\Project.mp4")
cap=cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_PLAIN
objects = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')
face_cascade=cv2.CascadeClassifier('C:\\Users\\Vishal\\cv2\\data\\haarcascade_frontalface_default.xml')
eyes_cascade=cv2.CascadeClassifier('C:\\Users\\Vishal\\cv2\\data\\haarcascade_eye.xml')
while True:
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face=face_cascade.detectMultiScale(gray,1.5,3)
    #eyes=eyes_cascade.detectMultiScale(gray,1.05,7)
    
    for (x,y,w,h) in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1)
        img=cv2.resize(gray[y:y+w+10,x:x+h+10],(48,48))
        arr= image.img_to_array(img)
        arr = np.expand_dims(arr, axis = 0)
        arr/= 255
        custom = model1.predict(arr)
        if objects[np.argmax(custom)]=='happy':
            cv2.putText(frame, str('Positive_expresion'), (x,y), font, 2, (255,0,0), 2)
        if objects[np.argmax(custom)]=='neutral':
            cv2.putText(frame, str('neutral_expression'), (x,y), font, 2, (255,0,0), 2 )
        if objects[np.argmax(custom)] in ['angry', 'disgust', 'fear', 'sad']:
            cv2.putText(frame, str('Negative_expression'), (x,y), font, 2, (255,0,0), 2 )
        eyes=eyes_cascade.detectMultiScale(gray[y:y+w+10,x:x+h+10],1.5,2)
        #print(eyes)
        #if :
              #  cv2.putText(frame, str('Negative_'), (x,y), font, 2, (255,0,0), 2)
        
        #for (x1,y1,w1,h1) in eyes:
        #    print(x1)
        
        if eyes==() :
             cv2.putText(frame, str('sleep'), (x,y-50), font, 2, (0,0,255), 2)
    
        if eyes!=():
            cv2.putText(frame, str('No_sleep'), (x,y-50), font, 2, (0,255,0), 2)
    frame=cv2.resize(frame,(1200,800))
    #cv2.imshow('face',gray[y:y+w+10,x:x+h+10])   
    cv2.imshow('video',frame)
    key=cv2.waitKey(1)
    if key ==ord('v'): 
        break 
cap.release()
cv2.destroyAllWindows()
```
